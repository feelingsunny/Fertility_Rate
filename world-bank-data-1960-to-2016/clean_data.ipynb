```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "fertility_rate.csv"

# Read Purchasing File and store into Pandas data frame
fertility_data = pd.read_csv(file_to_load)
fertility_data.head()
```

```python
fertility_data=fertility_data.drop(['Indicator Name','Indicator Code'], axis=1)
fertility_data.head()
```

```python
# File to Load 
file_to_load = "country_lat_long.csv"

# Read Purchasing File and store into Pandas data frame
country_data = pd.read_csv(file_to_load, encoding='cp1252')
country_data.head()
```

```python
country_data.rename(columns=
                    {'country':'Country Name'},
                    inplace = True
                )
country_data.head()
```

```python
fertility_country = pd.merge(left=fertility_data,right=country_data, how='left', left_on='Country Name', right_on='Country Name')
fertility_country
```

```python
fertility_country.dropna(axis=0, how='any')
```

```python
# File to Load 
file_to_load = "country_continent.csv"

# Read Purchasing File and store into Pandas data frame
region_data = pd.read_csv(file_to_load)
region_data.head()
```

```python
fertility_country_region = pd.merge(left=fertility_country,right=region_data, how='left', left_on='Country Name', right_on='Country Name')
fertility_country_region.head()
```

```python
fertility_df=fertility_country_region[['Country Name','2016','Country Code','region','latitude','longitude']]
fertility_df.head()
```

```python
fertility_df.dropna(axis=0, how='any')
fertility_df.head()
```

```python
fertility_df.index = np.arange(1, len(fertility_df) + 1)
fertility_df.head()
```

```python
fertility_df.index.name = 'New ID'
fertility_df.head()
```

```python
fertility_df.to_csv('data.csv',header=True)
```
